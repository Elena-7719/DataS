```python
def count_vowels(text):
    vowels = "aeiouAEIOUаеёиоуыэюяАЕЁИОУЫЭЮЯ"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count

user_input = input("Введите строку: ")
vowel_count = count_vowels(user_input)
print(f"Количество гласных букв: {vowel_count}")
```
